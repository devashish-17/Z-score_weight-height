# Mean with (+/-) 1,2,3 SD

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import zscore
```

```python
df = pd.read_csv('hs_heights_pair.csv')
df.head()
```

```python
boys_heights = df['boys']
girls_heights = df['girls']

mean_boys = np.mean(boys_heights)
std_boys = np.std(boys_heights)

mean_girls = np.mean(girls_heights)
std_girls = np.std(girls_heights)

print(f'Boys : Mean = {mean_boys}, Std_Dev = {std_boys}')
print(f'Girls : Mean = {mean_girls}, Std_Dev = {std_girls}')
```

```python
colors = ['green', 'orange', 'red']
```

```python
# Plot for Boys
plt.figure(figsize = (14, 6))

plt.subplot(1,2,2)
sns.histplot(boys_heights, kde=True, stat='density', linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_boys, std_boys)
plt.plot(x, p, 'k', linewidth=2)
title = f"Boys' Height\nMean : {mean_boys:.2f}; SD : {std_boys:.2f}"
plt.title(title)

# Plot mean and Std Dev lines for boys
for i in range(1, 4):
    plt.axvline(mean_boys + i * std_boys, color=colors[i-1], linestyle='--')
    plt.axvline(mean_boys - i * std_boys, color=colors[i-1], linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# Plot for Girls
plt.figure(figsize = (14, 6))

plt.subplot(1,2,2)
sns.histplot(girls_heights, kde=True, stat='density', linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_girls, std_girls)
plt.plot(x, p, 'k', linewidth=2)
title = f"Girls' Height\nMean : {mean_girls:.2f}; SD : {std_girls:.2f}"
plt.title(title)

# Plot mean and Std Dev lines for girls
for i in range(1, 4):
    plt.axvline(mean_girls + i * std_girls, color=colors[i-1], linestyle='--')
    plt.axvline(mean_girls - i * std_girls, color=colors[i-1], linestyle='--')

plt.tight_layout()
plt.show()
```

# Z-score
#### If data point is at some place, what is the area under that curve till that point

```python
# Convert Z-score to original values
z_score_boy = -1.79
z_score_girl = 0.48
boy_height = mean_boys + z_score_boy * std_boys
girl_height = mean_girls + z_score_girl * std_girls
print(boy_height, ' ', girl_height)
```

```python
# Plot the Z-score data point for the boy

# Plot for Boys
plt.figure(figsize = (14, 6))

plt.subplot(1,2,2)
sns.histplot(boys_heights, kde=True, stat='density', linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_boys, std_boys)
plt.plot(x, p, 'k', linewidth=2)
title = f"Boys' Height\nMean : {mean_boys:.2f}; SD : {std_boys:.2f}"
plt.title(title)

# Plot mean and Std Dev lines for boys
for i in range(1, 4):
    plt.axvline(mean_boys + i * std_boys, color=colors[i-1], linestyle='--')
    plt.axvline(mean_boys - i * std_boys, color=colors[i-1], linestyle='--')

plt.plot(boy_height, 0, 'o', color='darkred', markersize=10)

x_fill = np.linspace(xmin, boy_height, 100)
p_fill = norm.pdf(x_fill, mean_boys, std_boys)
plt.fill_between(x_fill, 0, p_fill, where=(x_fill <= boy_height), color='yellow', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Plot for Girls
plt.figure(figsize = (14, 6))

plt.subplot(1,2,2)
sns.histplot(girls_heights, kde=True, stat='density', linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_girls, std_girls)
plt.plot(x, p, 'k', linewidth=2)
title = f"Girls' Height\nMean : {mean_girls:.2f}; SD : {std_girls:.2f}"
plt.title(title)

# Plot mean and Std Dev lines for girls
for i in range(1, 4):
    plt.axvline(mean_girls + i * std_girls, color=colors[i-1], linestyle='--')
    plt.axvline(mean_girls - i * std_girls, color=colors[i-1], linestyle='--')

plt.plot(girl_height, 0, 'o', color='darkred', markersize=10)

x_fill = np.linspace(xmin, girl_height, 100)
p_fill = norm.pdf(x_fill, mean_girls, std_girls)
plt.fill_between(x_fill, 0, p_fill, where=(x_fill <= girl_height), color='yellow', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Z-score to Percentile

```python
df = pd.read_csv('weight-height.csv')
df.sample(5)
```

```python
w_mean = df['Weight'].mean()
w_std = df['Weight'].std()

h_mean = df['Height'].mean()
h_std = df['Height'].std()
```

```python
# Calculate the z-score for Weight and Height
df['Weight_Zscore'] = zscore(df['Weight'])
df['Height_Zscore'] = zscore(df['Height'])

print(f'Mean of Weight : {w_mean : .2f}')
print(f'SD of Weight   : {w_std : .2f}')
print(f'Mean of Height : {h_mean : .2f}')
print(f'SD of Height   : {h_std : .2f}')

df.head()
```

```python
# Save the modified DF to CSV
df.to_csv('height-weigh-zscore.csv')
```
